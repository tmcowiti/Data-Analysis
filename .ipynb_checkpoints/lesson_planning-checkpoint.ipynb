```python
import sys
import csv

filename = 'data/bibliography.csv'   #sys.argv[1]

raw = open(filename,'r')
file_reader = csv.reader(raw)

for line in file_reader:
    record = line
    #print record
    begin = record[0:3]
    authors = record[3]
    end = record[4:]
    for author in authors.split('; '):
        print begin, author, end
        
raw.close()
```
